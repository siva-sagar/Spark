#DF with | seperator

```
%fs head /FileStore/tables/july2023data/departments
```

```
dept_df = (spark.read
           .format("csv")
           .option("header","true")
           .option("delimiter","|")
           .load("/FileStore/tables/july2023data/departments")
)

dept_df.show()
```

```
dept_df.rdd.toDebugString()
```
